```python
meta = pd.read_csv('./Batch 1/meta_batch1_formatted.csv')
```

```python
# Takes vid from pathIn and splits frames between start and end to store in pathOut
def extract_frames(pathIn, pathOut, start, end, vid_name):
    

    cap = cv2.VideoCapture(pathIn)
    count = 0

    while (cap.isOpened()):

        ret, frame = cap.read()

        if ret == True:
            
            if (count in range(start, end)):
                print(vid_name+'_'+str(count))
                cv2.imwrite(os.path.join(pathOut, str(vid_name)+"_frame{:d}.jpg".format(count)), frame)  
            count += 1
        else:
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
vids = os.listdir('./Batch 1')[:-1]
```

```python
for each in vids[3:]:
  
  path = "./Batch 1/" + each
  
  for index, row in meta[meta['video'] == each].iterrows():

    if (row['label'] == 'walking') & (row['video'] == each):
        extract_frames(path, './Walking', row['frame_start'], row['frame_end'], each)

    if (row['label'] == 'falling') & (row['video'] == each):
        extract_frames(path, './Falling', row['frame_start'], row['frame_end'], each) 

    if (row['label'] == 'lying') & (row['video'] == each):
        extract_frames(path, './Lying', row['frame_start'], row['frame_end'], each) 

    if (row['label'] == 'sitting') & (row['video'] == each):
        extract_frames(path, './Sitting', row['frame_start'], row['frame_end'], each)  

    if (row['label'] == 'sleeping') & (row['video'] == each):
        extract_frames(path, './Sleeping', row['frame_start'], row['frame_end'], each) 
```
